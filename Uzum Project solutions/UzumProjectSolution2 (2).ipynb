```python
# 1. Data Loading and Exploration
import pandas as pd

# Load datasets
return_reasons_df = pd.read_parquet("return_reasons.parquet")
reviews_df = pd.read_parquet("reviews.parquet")
returns_df = pd.read_parquet("returns.parquet")
products_df = pd.read_parquet("products.parquet")
test_df = pd.read_parquet("test.parquet")

# Explore dataset contents and structure
print("Return Reasons:")
print(return_reasons_df.head())
print("\nReviews:")
print(reviews_df.head())
print("\nReturns:")
print(returns_df.head())
print("\nProducts:")
print(products_df.head())
print("\nTest Data:")
print(test_df.head())
```

```python
merged_data = pd.merge(reviews_df, returns_df, on="product_id", how="inner")
merged_data.head()
merged_data = pd.merge(merged_data, products_df, on="product_id", how="inner")
```

```python
merged_data.head()
```

```python
# Extract numerical features
num_cols = ['rating', 'purchase_price']  # Define numerical columns
X_features = merged_data[num_cols].values

# Define target variable
y = merged_data["cause"]
```

```python
# 3. Model Training
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

# Split data into train and validation sets
X_train, X_val, y_train, y_val = train_test_split(X_features, y, test_size=0.2, random_state=42)

# Initialize and train the model
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)
```

```python
# 4. Model Evaluation
from sklearn.metrics import f1_score

# Predict on validation set
y_val_pred = model.predict(X_val)

# Calculate F1-score
f1_macro = f1_score(y_val, y_val_pred, average='macro')
print("Validation F1-score:", f1_macro)
```

```python
# 5. Prediction
# Load test dataset
test_data = pd.merge(test_df, products_df, on="product_id", how="inner")
X_test_features = test_data[num_cols].values

# Make predictions on test data
y_test_pred_prob = model.predict_proba(X_test_features)
```

```python
# 6. Output Preparation
# Map predicted probabilities to reasons using return_reasons_df
predicted_reasons = return_reasons_df['reason'].values

# Create DataFrame with probabilities and reasons
result_df = pd.DataFrame({
    'product_id': test_df['product_id'],
    'order_item_id': test_df['order_item_id'],
    'prob_return_reason_DEFECTED': y_test_pred_prob[:, 0],
    'prob_return_reason_WRONG_ITEM': y_test_pred_prob[:, 1],
    'prob_return_reason_BAD_QUALITY': y_test_pred_prob[:, 2],
    'prob_return_reason_PHOTO_MISMATCH': y_test_pred_prob[:, 3],
    'prob_return_reason_WRONG_SIZE': y_test_pred_prob[:, 4]
})

# Save to parquet
result_df.to_parquet("result.parquet", index=False)
```
